```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
%matplotlib inline
import seaborn as sns
```

```python
df0 = pd.read_csv("Cluster0.csv",index_col=0,parse_dates=True)
```

```python
from statsmodels.tsa.seasonal import seasonal_decompose

from pylab import rcParams
rcParams['figure.figsize'] = 12,8

result = seasonal_decompose(df0['y'], model='multiplicative')  # model='mul' also works
result.plot();
```

### SMA

```python
df0['6-month-SMA'] = df0['y'].rolling(window=6).mean()
df0['12-month-SMA'] = df0['y'].rolling(window=12).mean()
```
